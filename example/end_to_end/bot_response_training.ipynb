# Import

```python
from query_insights.api import QueryInsights, get_bot_error_message
import random
import pandas as pd
```

# Input

Config path

```python

# questions = ["Get total shipped quantity by source location.",
#              "What are the top 3 high selling brands"]

questions = pd.read_excel("../../data/output_folder/questions_bot_training.xlsx").values.ravel()
language = "english"
len(questions)
```

```python
data_config_path = "../../configs/local/data_config.yaml"
user_config_path = "../../configs/local/user_config.yaml"
model_config_path = "../../configs/model_config.yaml"
```

# Extract Insights

```python
qi = QueryInsights(
    user_config_path=user_config_path,
    data_config_path=data_config_path,
    model_config_path=model_config_path,
    logging_level="DEBUG",
)
```

## Track 1

```python
for question in questions:
    track1_response = qi.text_to_query(question=question, additional_context=None, language=language)

    if track1_response["status"] == "success":
        track1_query, track1_table, track1_dict = track1_response["output"]
        bot_response = random.choice(
            [
                "Done! Here are the results.",
                "All set! Here's what I found.",
                "Got it! Here are the results for you.",
                "Finished! Here's what you were looking for.",
                "Here's what I came up with.",
                "Completed! Here's what I found out for you.",
                "Task accomplished! Here are the results you requested.",
                "Task fulfilled! Here's what I found.",
                "Done and done! Here are the results you asked for.",
            ]
        )
        
    elif track1_response["status"] == "skip":
        track1_query, track1_table = track1_response["output"]
        bot_response = "Track1 skipped"
        
    elif track1_response["status"] == "failure":
        error_message, track1_query = track1_response["output"]
        track1_query = (
            error_message + " \nFollowing is the generated query: \n\n" + track1_query
        )
        bot_response = get_bot_error_message(track1_query)

    qi.create_bot_training_data([[question, bot_response]])
```
